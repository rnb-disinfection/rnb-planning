```python
from sklearn import svm

self = rtrain
ROBOT_TYPE=RobotType.indy7
save_model=False
C_svm = 170

self.featurevec_list_train, self.success_list_train = self.load_data(ROBOT_TYPE, "train")
self.featurevec_list_test, self.success_list_test = self.load_data(ROBOT_TYPE, "test")
train_idx = [i_f for i_f, featurevec in enumerate(self.featurevec_list_train) if 0.2<featurevec[0]<1.5]
test_idx = [i_f for i_f, featurevec in enumerate(self.featurevec_list_test) if 0.2<featurevec[0]<1.5][:5000]

feature_mat_train = np.array(
    [(radius, theta, height, azimuth_loc, zenith, radius**2, np.linalg.norm([radius, height-0.33]), np.linalg.norm([radius, height-0.33])**2) 
     for radius, theta, height, azimuth_loc, zenith in self.featurevec_list_train])[train_idx]
feature_mat_test = np.array(
    [(radius, theta, height, azimuth_loc, zenith, radius**2, np.linalg.norm([radius, height-0.33]), np.linalg.norm([radius, height-0.33])**2) 
     for radius, theta, height, azimuth_loc, zenith in self.featurevec_list_test])[test_idx]
success_list_train_cur = np.array(self.success_list_train)[train_idx]
success_list_test_cur = np.array(self.success_list_test)[test_idx]

self.clf = svm.SVC(kernel='rbf', C=C_svm)
self.clf.fit(feature_mat_train, success_list_train_cur)
if save_model:
    self.save_model()

gtimer = GlobalTimer.instance()
gtimer.reset()
with gtimer.block("trainset"):
    train_res = np.equal(self.clf.predict(feature_mat_train), success_list_train_cur)
with gtimer.block("testset"):
    test_res = np.equal(self.clf.predict(feature_mat_test), success_list_test_cur)
print("=" * 80)
print(gtimer)

print("{}".format(C_svm)+"=" * 80)
print("trainning accuracy = {} %".format(round(np.mean(train_res) * 100, 2)))
print("test accuracy = {} %".format(round(np.mean(test_res) * 100, 2)))
# print("=" * 80)
# print("trainning success accuracy = {} %".format(
#     round(np.mean(train_res[np.where(success_list_train_cur)]) * 100, 2)))
# print("trainning failure accuracy = {} %".format(
#     round(np.mean(train_res[np.where(np.logical_not(success_list_train_cur))]) * 100, 2)))
# print("=" * 80)
# print("test success accuracy = {} %".format(round(np.mean(test_res[np.where(success_list_test_cur)]) * 100, 2)))
# print("test failure accuracy = {} %".format(
#     round(np.mean(test_res[np.where(np.logical_not(success_list_test_cur))]) * 100, 2)))
# print("=" * 80)
```
